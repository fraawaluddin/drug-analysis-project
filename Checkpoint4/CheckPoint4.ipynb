## Outliers Values

```python
results = []

cols = Datafile.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = Datafile[col].quantile(0.25)
  q3 = Datafile[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = Datafile[(Datafile[col] < lower_bound) | (Datafile[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(Datafile))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_Datafile = pd.DataFrame(results)
results_Datafile.set_index('Kolom', inplace=True)
results_Datafile = results_Datafile.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_Datafile)
```

Berdasarkan metode IQR diatas bisa dilihat bahwa data yang memiliki persentase outliers hanya Improvement_Score sebesar 0.7%, hal tersebut tergolong cukup kecil. Bisa dibilang dataset ini cukup stabil

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=Datafile['Improvement_Score'])
plt.title('Distribusi dan Deteksi Outlier pada Improvement Score')
plt.ylabel('Improvement Score')
plt.show()
```

Jika kita lihat pada boxplot diatas terdapat outlier yang wajar karena membahas terkait indikator penyembuhan pasien

## Construct Data

```python
def categorize_improvement(Improvement_Score):
    if Improvement_Score <= 5:
        return "Low"
    elif Improvement_Score <= 7.5:
        return "Medium"
    else:
        return "High"

Datafile["Improvement_Level"] = Datafile["Improvement_Score"].apply(categorize_improvement)
Datafile.head()
```

Untuk mempermudah pembacaan data, diperlukan interpretasi tingkat efektivitas pengobatan dengan mengubah data numerik Improvement_Score menjadi kategori tingkat penyembuhan pasien. Berdasarkan kebutuhan dataset, saya membutuhkan kolom Improvement_Score yang memiliki indikator low pada score <= 5, medium pada score 5-7.5, dan high dengan score di atas 7.5. Berdasarkan kode di atas, saya berhasil membuat kolom baru yaitu Improvement_Level yang mengelompokkan pasien berdasarkan tingkat penyembuhan pasien. Hal tersebut untuk mempermudah analisis dan bisa mendapatkan kesimpulan dari faktor yang berpengaruh terhadap tingkat penyembuhan pasien dan juga mempermudah perbandingan kelompok pasien terhadap tingkat penyembuhan pasien.

## BoxPlot

### Distribusi Improvement Score berdasarkan Top 5 Jenis Obat

Pada boxplot dibawah, saya ingin melihat distribusi tingkat perbaikan pasien pada setiap jenis obat. Tujuannya untuk mengetahui efektivitas obat berdasarkan variasi.

Berdasarkan boxplot distribusi improvement score pada 5 jenis obat teratas, bisa dilihat bahwa kelima jenis obat tersebut memiliki median yang hampir sama, yang terdapat pada kisaran 6 sampai 7.5.
