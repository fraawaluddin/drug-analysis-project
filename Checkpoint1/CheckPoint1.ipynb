#Business Understanding

##Bussiness Objective

Dataset Drugs and Side Effects berisi informasi pasien yang
menjalani pengobatan dengan berbagai jenis obat, termasuk
usia pasien, jenis kelamin, kondisi penyakit, dosis obat,
durasi pengobatan, efek samping, serta tingkat perbaikan
kondisi pasien (Improvement Score).

Permasalahan utama yang ingin dipahami adalah bagaimana
efektivitas penggunaan obat terhadap tingkat perbaikan pasien.
Selain itu, penting untuk mengetahui apakah faktor seperti
dosis obat, durasi pengobatan, dan efek samping memiliki
pengaruh terhadap hasil pengobatan yang diperoleh pasien.

##Asses Situation

Dataset diperoleh dari platform Kaggle dan berisi data pasien
dengan berbagai kondisi medis dan jenis pengobatan. Dataset
memiliki beberapa variabel kategorikal dan numerik yang dapat
digunakan untuk analisis efektivitas obat.

Beberapa tantangan yang mungkin terdapat pada dataset ini
antara lain variasi respons pasien terhadap obat yang berbeda,
kemungkinan adanya ketidakseimbangan jumlah data pada setiap
jenis obat atau kondisi penyakit, serta perlunya pengecekan
kualitas data seperti missing value atau inkonsistensi data
sebelum dilakukan analisis lebih lanjut.

##Analytic Goals

Untuk mengatasi permasalahan tersebut, analisis akan dilakukan
dengan tujuan:

1. Mengidentifikasi jenis obat yang memberikan tingkat
   improvement score tertinggi.
2. Menganalisis hubungan antara dosis obat dan tingkat
   perbaikan pasien.
3. Menganalisis pengaruh durasi pengobatan terhadap hasil
   perawatan pasien.
4. Mengidentifikasi efek samping yang paling sering muncul.
5. Membandingkan tingkat perbaikan pasien berdasarkan kondisi
   penyakit dan gender.
