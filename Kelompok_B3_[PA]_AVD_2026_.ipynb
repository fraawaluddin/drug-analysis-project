## **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Pada tahap ini dilakukan pengecekan outliers values menggunakan metode IQR(Interquartile Range), dimana ditemukan outlier pada seluruh kolom numerik. Waste (kg) memiliki persentase outlier tertinggi sebesar 8%, diikuti oleh Energy (kWh) 7,48%, Water (liters) 6,48%, Defect Rate 3,9%, dan Recycled Ratio 1,6%. Nilai-nilai ekstrem ini berpotensi memengaruhi hasil analisis statistik, sehingga perlu ditangani pada tahap Data Preparation.

## **Outliers Values**

```python
columns_to_impute = ["Production Volume", "Energy (kWh)", "Water (liters)", "Recycled Ratio", "Defect Rate"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Pada tahap ini dilakukan penanganan outlier menggunakan metode IQR. prosesnya adalah sebagai berikut:
- Q1 = nilai persentil ke-25 (batas bawah distribusi normal)
- Q3 = nilai persentil ke-75 (batas atas distribusi normal)
- IQR = Q3 - Q1 (rentang antar quartil)
Lower bound = Q1 - 1.5 x IQR (batas bawah wajar)
Upper bound = Q3 + 1.5 x IQR (batas atas wajar)
Nilai yang berada di bawah lower bound akan dinaikkan ke lower bound, dan nilai yang melebihi upper bound akan diturunkan ke upper bound. Metode ini dipilih karena tidak menghapus data, melainkan hanya membatasi nilai ekstrem ke rentang yang wajar.

```python
columns_to_impute = ["Waste (kg)"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    upper_bound = Q3 + 1.5 * IQR

    if col == "Waste (kg)":
        df.loc[:, col] = df[col].clip(lower=0, upper=upper_bound)
    else:
        lower_bound = Q1 - 1.5 * IQR
        df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Karena Waste (kg) memiliki pertimbangan logika bisnis yang khusus dan limbah tidak mungkin bernilai negatif. Jika menggunakan rumus IQR biasa, lower bound bisa menghasilkan nilai negatif seperti -50 kg yang tidak masuk akal.
Maka dari itu, lower bound diganti langsung menjadi 0 yang artinya nilai negatif apapun akan dinaikkan ke 0, sementara upper bound tetap menggunakan IQR normal untuk menangani nilai yang terlalu tinggi.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Setelah proses dilakukan, dilakukan pengecekan ulang menggunakan metode IQR yang sama untuk memastikan tidak ada outlier yang tersisa. Hasilnya menunjukkan bahwa seluruh kolom berhasil mencapai 0.0% outlier.

# **Construct Data**

```python
df['Efficiency'] = df['Waste (kg)'] / df['Production Volume']
```

Dibuat kolom baru bernama Efficiency yang merpakan hasil pembagian antara Waste (kg) dengan Production Volume. Kolom ini mempresentasikan jumlah limbah yang dihasilkan per satu unit produksi. Semakin kecil nilai Efficiency, semakin efisien proses produksi karena limbah yang dihasilkan lebih sedikit per unit produk.

```python
def kategori_efisiensi(x):
    if x < 0.097:
        return 'Efisien'
    elif x < 0.229:
        return 'Cukup Efisien'
    else:
        return 'Tidak Efisien'
```

Dibuat fungsi kategori_efisiensi untuk mengkategorikan nilai Efficiency ke dalam 3 kategori:
- Efisien = nilai Efficiency < 0.097 (limbah < 0.097 kg per unit produksi)
- Cukup Efisien = nilai Efficiency < 0.229
- Tidak Efisien = nilai Efficiency >= 0.229 (limbah >= 0.229 kg per unit produksi)

```python
df['Efficiency Category'] = df['Efficiency'].apply(kategori_efisiensi)
```

Fungsi kategori_efisiensi diterapkan ke seluruh kolom Efficiency dan menghasilkan kolom baru Efficiency Category.
