## Outliers Values

```python
results = []

cols = Datafile.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = Datafile[col].quantile(0.25)
  q3 = Datafile[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = Datafile[(Datafile[col] < lower_bound) | (Datafile[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(Datafile))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_Datafile = pd.DataFrame(results)
results_Datafile.set_index('Kolom', inplace=True)
results_Datafile = results_Datafile.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_Datafile)
```

Berdasarkan metode IQR diatas bisa dilihat bahwa data yang memiliki persentase outliers hanya Improvement_Score sebesar 0.7%, hal tersebut tergolong cukup kecil. Bisa dibilang dataset ini cukup stabil

# Eksplorasi Data (EDA)

## Comparison(Perbandingan)

Aktivitas: Membandingkan rata-rata Improvement_Score berdasarkan Drug_Name, Gender, dan Condition.

Tujuan: Mengidentifikasi obat dengan tingkat perbaikan tertinggi serta mengetahui apakah terdapat perbedaan tingkat perbaikan pasien berdasarkan gender dan kondisi penyakit.

Visualisasi: Bar Chart (Grafik Batang).

```python
avg_improvement = Datafile.groupby("Drug_Name")["Improvement_Score"].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=avg_improvement.index, y=avg_improvement.values, hue=avg_improvement.index, palette="viridis")
plt.title("Rata-rata Improvement Score per Drug")
plt.xlabel("Drug Name")
plt.ylabel("Average Improvement Score")
plt.xticks(rotation=45)
plt.show()
```

Berdasarkan grafik perbandingan rata-rata Improvement Score per Drug,
terlihat bahwa seluruh obat memiliki nilai rata-rata yang relatif
berdekatan, yaitu berada pada kisaran sekitar 6,8 hingga 7,3.
Bisa ditarik kesimpulan bahwa seluruh obat tidak memiliki perbedaan signifikan pada rata rata improvement score per drug, dan juga tidak ada drug yang menonjol sendiri improvement scorenya.

```python
avg_gender = Datafile.groupby("Gender")["Improvement_Score"].mean().sort_values(ascending=False)

plt.figure(figsize=(6,5))
sns.barplot(x=avg_gender.index, y=avg_gender.values, hue=avg_gender.index, palette="viridis")
plt.title("Rata-rata Improvement Score berdasarkan Gender")
plt.xlabel("Gender")
plt.ylabel("Average Improvement Score")
plt.xticks(rotation=45)
plt.show()
```

Berdasarkan grafik rata-rata Improvement Score berdasarkan gender, terlihat bahwa nilai rata-rata antara pasien laki-laki dan perempuan hampir sama yaitu sekitar 7. Artinya tidak terdapat perbedaan signifikan dalam tingkat perbaikan pasien berdasarkan gender. Jadi, efektivitas pengobatan dalam dataset konsisten untuk kedua gender bukan merupakan faktor dominan dalam menentukan keberhasilan improvement score/terapi.

```python
avg_condition = Datafile.groupby("Condition")["Improvement_Score"].mean().sort_values(ascending=False)

plt.figure(figsize=(8,6))
sns.barplot(x=avg_condition.index, y=avg_condition.values, hue=avg_condition.index, palette="viridis")
plt.title("Rata-rata Improvement Score berdasarkan Condition")
plt.xlabel("Condition")
plt.ylabel("Average Improvement Score")
plt.xticks(rotation=45)
plt.show()
```

Berdasarkan grafik rata-rata Improvement Score berdasarkan kondisi penyakit, terlihat bahwa seluruh kondisi memiliki nilai rata-rata yang relatif berdekatan, berada pada kisaran sekitar 7. artinya efektivitas pengobatan dalam dataset ini cenderung konsisten di berbagai kondisi penyakit. Tidak ada perbedaan signifikan antar kondisi, jadi jenis kondisi bukan faktor dominan dalam menentukan tingkat perbaikan pasien.

## Composition(Komposisi)

Aktivitas: Melihat proporsi atau persentase efek samping yang paling sering muncul pada pasien.

Tujuan: Mengetahui efek samping yang paling dominan dalam penggunaan obat serta memahami risiko yang paling umum terjadi.

Visualisasi: Pie Chart (Grafik Lingkaran).

```python
side_effect_counts = Datafile.groupby("Side_Effects")["Patient_ID"].count().sort_values(ascending=False)
side_effect_top5 = side_effect_counts.head(5)

plt.figure(figsize=(10,8))
side_effect_top5.plot(kind="pie", autopct="%1.1f%%", startangle=140)
plt.title("Proporsi 5 Efek Samping Terbanyak")
plt.ylabel("")
plt.axis("equal")
plt.show()
```

Berdasarkan grafik diatas efek samping terbanyak yaitu Nausea dengan persentase 38.2%, lalu yang kedua Dizziness dengan persentase 22.4%, ketiga Headache dengan persentase 17.6%, diikuti Fatigue 11%, yang terakhir Rash dengan persentase 10.8%. Hal ini menunjukkan bahwa mual adalah risiko utama pada efek samping obat. Hal ini dapat menjadi pertimbangan dalam pemilihan obat alternatif dengan profil efek samping yang lebih ringan.

## Distribution(Distribusi)

Aktivitas: Menganalisis penyebaran nilai Improvement_Score pada seluruh pasien.

Tujuan: Melihat apakah tingkat perbaikan pasien cenderung stabil atau memiliki variasi yang tinggi, dan melihat apakah ada distribusi ekstrem dalam data.

Visualisasi: Histogram.

```python
plt.figure(figsize=(10,6))
sns.histplot(Datafile["Improvement_Score"], bins=20, kde=True)
plt.title("Distribusi Improvement Score Pasien")
plt.xlabel("Improvement Score")
plt.ylabel("Frekuensi")
plt.show()
```

Berdasarkan histogram distribusi Improvement Score, kebanyakan pasien memiliki Improvement_Score pada rentang 6-8, yang artinya kebanyakan pasien mengalami tingkat perbaikan yang cukup baik setelah pengobatan.

Artinya efektivitas obat dalam dataset cenderung stabil dan konsisten pada sebagian besar pasien, dan tidak ada dominasi nilai ekstrem yang signifikan.

## Relationship/Hubungan

Aktivitas: Menganalisis hubungan antara variabel numerik (Dosage_mg dan Treatment_Duration_days) dengan Improvement_Score.

Tujuan: Mengetahui apakah peningkatan dosis atau durasi pengobatan berbanding lurus dengan peningkatan tingkat perbaikan pasien.

Visualisasi: Heatmap.

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    Datafile[["Dosage_mg", "Treatment_Duration_days", "Improvement_Score"]].corr(),
    annot=True,
    cmap="viridis",
    fmt=".2f")
plt.title("Korelasi antar Variabel Numerik")
plt.show()
```

Jika melihat heatmap di atas, nilai korelasi antara Dosage_mg dan Improvement_Score sebesar 0.04 serta antara Treatment_Duration_days dan Improvement_Score sebesar 0.05, yang berarti hubungannya positif tetapi sangat lemah. Jadi, bisa disimpulkan bahwa Dosage_mg tidak secara signifikan meningkatkan Improvement_Score dan peningkatan dosis tidak dapat dijadikan strategi utama untuk meningkatkan efektivitas pengobatan, serta Treatment_Duration_days juga tidak signifikan meningkatkan Improvement_Score, sehingga memperpanjang masa pengobatan belum tentu menghasilkan peningkatan hasil perbaikan pasien/terapi pasien.
